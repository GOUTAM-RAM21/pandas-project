```python
import pandas as pd

# Example data as Python dictionary
data = {
    "Name": ["Goutam", "Ravi", "Neha", "Pooja"],
    "Age": [23, 21, 25, 22],
    "City": ["Jodhpur", "Delhi", "Mumbai", "Jaipur"]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Show the full DataFrame
print("Full DataFrame:")
print(df)

# Access a single column
print("\nNames column:")
print(df["Name"])

# Basic statistics
print("\nStatistics:")
print(df.describe())

# Filter rows where Age > 22
print("\nPeople older than 22:")
print(df[df["Age"] > 22])
```
