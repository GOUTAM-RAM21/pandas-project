```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

dataset = pd.read_csv("Insurance.csv")
print(dataset)
```

```python
dataset['bought_insurance'].replace({'no':'0','yes':'1'},inplace=True)
dataset
```

```python
plt.scatter(x='age',y='bought_insurance',data='dataset')
```

```python

plt.figure(figsize=(8,6))
plt.scatter(x='age', y='bought_insurance', data=dataset)
plt.show()

```

```python
from sklearn.model_selection import train_test_split

# Splitting the data
x_train, x_test, y_train, y_test = train_test_split(
    dataset[['age']],  # Feature(s) (age)
    dataset['bought_insurance'],  # Target variable
    test_size=0.2  # 20% of the data will be used for testing
)

# Checking the length of the training and testing data
print(len(x_train))  # Length of training data
print(len(x_test))   # Length of test data
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression

# Create an instance of Logistic Regression
lr = LogisticRegression()

# Fit the model to the training data
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test)
```
